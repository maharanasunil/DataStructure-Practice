## Linked List(Deleting a Node)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class Linkedlist:
    def __init__(self):
        self.head = None
        
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
    
    #Given a reference to the head of the LL anda 'key'
    #delete the first occurence of the key in LL
    def deleteNode(self, key):
        #store the head node
        temp = self.head
        #if the value of head node matches with the key then delete it
        if (temp is not None):
            if (temp.data == key):
                self.head = temp.next
                temp = None
                return
            
        #Search for the key to be deleted also keep the track of previous node
        #As we have to change the next of previous node
        while(temp is not None):
            if(temp.data == key):
                break
            prev = temp
            temp = temp.next
            
        #If the key was not present in LL
        if(temp == None):
            return "Empty L.L"
        
        #Unlink the node from L.L
        prev.next = temp.next
        temp = None
        
        
    #Function to print L.L
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data, end = '->')
            temp = temp.next
                
#Driver Code
llist = Linkedlist()
llist.push(4)
llist.push(3)
llist.push(2)
llist.push(1)

print('Created Linked List :\n')
llist.printList()

llist.deleteNode(1)
print('\n\nLinked List After deleting Node1 :\n')
llist.printList()
```

## Delete a Linked List node at a given position

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
class LinkedList:
    def __init__(self):
        self.head = None
        
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
        
    #Given a reference to head of L.L and a position
    #We have to delete the node at given position
    def deleteNode(self, pos):
        #if LL is empty
        if(self.head == None):
            return
        
        #store head node
        temp = self.head
        
        #if the head needs to be deleted
        if pos == 0:
            self.head = temp.next
            temp = None
            return
        
        #Find previous node - of the node to be deleted
        for _ in range(pos - 1):
            temp = temp.next
            if temp == None:
                break
                
        #If the pos is more than no of nodes
        if temp is None:
            return
        if temp.next is None:
            return
            
        #Node temp.next is the node which is to be deleted
        #Store the pointer to the next - of the node to be deleted
        next = temp.next.next
        
        #Unlink node from LL
        temp.next = None
        temp.next = next
        
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data, end='->')
            temp = temp.next
            
#Driver Code
llist = LinkedList()
llist.push(5)
llist.push(4)
llist.push(3)
llist.push(2)
llist.push(1)
llist.push(0)
print("Created Linked List:\n")
llist.printList()

llist.deleteNode(3)

print("\n\nL.L After Deleting Node 3:\n")
llist.printList()
```
